Question 1: What is Logistic Regression, and how does it differ from Linear
Regression?

Ans:- Logistic regression is an classification algo which is used to predict a categorical outcome, most commonly a binary outcome 
        Logistic regression predicts the probability that an observation belongs to a particular class.
        
       

 Linear Regression:- it is a supervised learning algorithm used for predicting the continious values like temp,prices,salaries
                            In simple it tries to draw a best fit line to get the correct continious prediction.

Question 2: Explain the role of the Sigmoid function in Logistic Regression.

Ans:- Role of sigmoid function in logistic Regression
        sigmoid function converts the standard linear model output into a probability 
        it presents the output between 0 and 1 
        0 means very low chance of having class 1
        1 means very high chance of having the probability of class 1
