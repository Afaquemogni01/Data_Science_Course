Question 2: What is the difference between Bagging and Boosting?

Difference between Bagging and boosting

bagging:-Models are trained independently on different random subsets of training data ,their results are then combined usually by avgeraging (for regression) or voting (for classificaton),helps to reduce the varience and prevents overfitting

Boosting:- Ensemble teachnique that combines multiple week learners to create a strong learner
weak models are trained in series such that each next model tries to correct errors of the previous model until the entire the model training datasets predicted correctly

```python
data=fetch_california_housing()
```

```python
x,y=data.data,data.target
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
bagging=BaggingRegressor(
    estimator=DecisionTreeRegressor(),
    n_estimators=100,
    max_samples=0.8,
    bootstrap=True,
    random_state=42
)
```

```python
bagging.fit(x_train,y_train)
```

```python
bag_pred=bagging.predict(x_test)
```

```python
bagging_mse=mean_squared_error(y_test,bag_pred)
```

```python
bagging_mse
```

```python
rf=RandomForestRegressor(
    n_estimators=100,
    max_depth=None,
    random_state=42
)
```

```python
rf_pred=rf.predict(x_test)
```

```python
rf_mse=mean_squared_error(y_test,rf_pred)
```
