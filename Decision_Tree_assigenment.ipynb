Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision
Trees? Give one practical advantage of using each.

Pre-pruning:-Decision Tree growth is stoped before the tree becomes too complex
the alg sets the condition like:-
= Max tree depth
= Min samples required to split the node
= Min info  gain required for a split


Post Pruning :- The tree is first grown fully (allowing) all splits until the leaves are pure or meet the min samples 
After training branches with little predictive power are cut back

Question 6: Write a Python program to:

● Load the Iris Dataset

● Train a Decision Tree Classifier using the Gini criterion

● Print the model’s accuracy and feature importances

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
data
```

```python
df=pd.DataFrame(data=data.data,columns=data.feature_names)
```

```python
df['target']=data.target
```

```python
df
```

```python
# now separate the  x and y

x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# now train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
# now start the training of the model using the decision tree classifier 

from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='gini',random_state=42)
```

```python
classifier.fit(x_train,y_train)
```

```python
# now the model is trained 
# now print the models accuracy


y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
print(acc)
```

```python
print("Feature Importances:")
for name, importance in zip(data.feature_names, classifier.feature_importances_):
    print(f"{name}: {importance:.3f}")
```
